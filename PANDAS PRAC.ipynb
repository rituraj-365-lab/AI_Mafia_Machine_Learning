# Replace Using Mean, Median, or Mode #

```python
df4
```

```python
x=df4["Age"].median()
n3=df4.fillna({"Age":x})
n3
```

```python
y=df4["Marks"].median()
n3=df4.fillna({"Marks":y})
n3
```

```python
df4["Salary"] = pd.to_numeric(df4["Salary"], errors="coerce")
df4
```

```python
z=df4["Salary"].median()
n3=df4.fillna({"Salary":z})
n3
```
